```python

import numpy as np
import pandas as pd
df = pd.read_csv("Flats Uncleaned Dataset.csv")
df.head(10) 
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["transaction"].fillna(df["transaction"].mode()[0], inplace=True)
```

```python
df["description"].fillna(df["description"].mode()[0], inplace=True)
```

```python
df["floor"].fillna(df["floor"].mode()[0], inplace=True)
```

```python
df["price_per_sqft"].fillna(df["price_per_sqft"].mean(), inplace=True)

```

```python
df["status"].fillna(df["status"].mode()[0], inplace=True)
```

```python
df["furnishing"].fillna(df["furnishing"].mode()[0], inplace=True)
```

```python
df["facing"].fillna(df["facing"].mode()[0], inplace=True)
```

```python
import matplotlib.pyplot as plt
df2=df.sample(10)
plt.figure(figsize=(8,5))
plt.plot(df2['property_name'],df2["status"],marker='*',color='g')
plt.title("property")
plt.xlabel("property name")
plt.ylabel("status") 
plt.show()
```

```python

plt.figure(figsize=(8,5))
plt.bar(df2['price'],df2['status'])
plt.title("property")
plt.xlabel("property name")
plt.ylabel("status") 
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df2['property_name'],df2['status'])
plt.title("property")
plt.xlabel("property name")
plt.ylabel("status") 
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df2['status'], bins=10, color='skyblue', edgecolor='black' )
plt.title("Status Distribution")
plt.xlabel("Status")
plt.ylabel("Frequency")
plt.show()
```

```python
dep=df["transaction"].value_counts()
plt.figure(figsize=(8,5))
plt.pie(dep, labels=dep.index, autopct='%1.1f%%',)
plt.title("Department distribution of students")
plt.show()
```

```python

plt.figure(figsize=(8,5))
plt.plot(df2['property_name'],df2["status"],marker='*',color='g', label='Marks')
plt.plot(df2['property_name'],df2["transaction"],marker='o',color='b', label='Age')
plt.plot(df2['property_name'],df2["price"],marker='o',color='r', label='Attendance')
plt.title(" history of property")
plt.xlabel("Property Name")
plt.ylabel("Values") 
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
status_counts = df2['status'].value_counts()
plt.bar(status_counts.index, status_counts.values)
plt.title('Status counts')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
price_values = pd.to_numeric(df2['price'].astype(str).str.replace(r'[^0-9.]', '', regex=True), errors='coerce')
plt.plot(df2['property_name'], price_values, marker='o')
plt.title('Price by property')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
