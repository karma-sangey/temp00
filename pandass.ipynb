```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = 'students_data_2000_rows.csv'
df = pd.read_csv(file_path)

# 1. Students with more than 85 marks, grouped by city
students_above_85 = df[df['Marks'] > 85]
students_above_85_by_city = students_above_85.groupby('City').size().reset_index(name='student_count')

# 2. CSE students sorted by attendance
cse_students_by_attendance = df[df['Department'] == 'CSE'].sort_values('Attendance', ascending=False)

# 3. Students with age greater than 20 and less than 25
age_filter = df[(df['Age'] > 20) & (df['Age'] < 25)]

# 4. Add a new column 'Fees' with random paid/pending values
np.random.seed(42)
df['Fees'] = np.random.choice(['paid', 'pending'], size=len(df))

# 5. Add a new column 'Result' based on marks
conditions = [
    df['Marks'] > 90,
    (df['Marks'] >= 80) & (df['Marks'] <= 90),
    (df['Marks'] >= 70) & (df['Marks'] < 80)
]
choices = ['A grade', 'B grade', 'C grade']
df['Result'] = np.select(conditions, choices, default='P grade')

# 6. Group by Gender and Department, showing average attendance per department
avg_attendance_by_gender_department = df.groupby(['Gender', 'Department'])['Attendance'].mean().reset_index()

# Display results
print('1) Students with marks > 85 grouped by city:')
print(students_above_85_by_city.to_string(index=False))
print('\n2) CSE students by attendance:')
print(cse_students_by_attendance[['Name', 'Attendance']].to_string(index=False))
print('\n3) Students aged between 21 and 24:')
print(age_filter[['Name', 'Age', 'Department', 'City', 'Marks']].to_string(index=False))
print('\n4) DataFrame with Fees column:')
print(df[['Name', 'Fees']].head().to_string(index=False))
print('\n5) Result column preview:')
print(df[['Name', 'Marks', 'Result']].head().to_string(index=False))
print('\n6) Average attendance by gender and department:')
print(avg_attendance_by_gender_department.to_string(index=False))
```
